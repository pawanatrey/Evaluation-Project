```python
df[['Gender','Married','Self_Employed']]=df1
df.info()
```

```python
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].median())
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].mode().iloc[0])
df
```

```python
# Impute the null values of descreate data.

from sklearn.impute import SimpleImputer

si_imp=SimpleImputer(strategy='mean')
fil_null=si_imp.fit_transform(df[['LoanAmount','ApplicantIncome']])
df2=pd.DataFrame(fil_null)
df2
```

```python
# Split Numerical data and Categorical Data
# It will tell these are the categorical column in dataset

categorical_object=[]
for i in df.dtypes.index:
    if df.dtypes[i]=='object':
        categorical_object.append(i)      
print("Categorical Data",categorical_object)    
```

```python
# It will tell these are the numerical column in dataset

numerical_object=[]
for i in df.dtypes.index:
    if (df.dtypes[i]!='object'):
        numerical_object.append(i) 
print("Numerical Data\n",numerical_object)          
```

```python
# Here we check how data is distributed, it is normally distributed or not.

# After go through the dataset we assume something that skewness and outliers is there so now we analyse the data to check 
# howmuch our assumption is correct.

plt.figure(figsize=(20,15),facecolor='yellow')
number=1
for i in numerical_object:
    if number<=6:
        ax=plt.subplot(2,3,number)
        sns.distplot(df[i])
        plt.xlabel(i,fontsize=10)
    number+=1
plt.show()                                                                                       
```

```python
plt.figure(figsize=(20,15),facecolor='yellow')
number=1
for i in numerical_object:
    if number<=6:
        ax=plt.subplot(2,3,number)
        sns.boxplot(df[i])
        plt.xlabel(i,fontsize=10)
    number+=1
plt.show()
```

```python
# Find the standard number of outlier and remove it

from scipy.stats import zscore

z_score=zscore(df[['ApplicantIncome','CoapplicantIncome','LoanAmount']])
data=np.abs(z_score)
filter_data=(data<3).all(axis=1)
df=df[filter_data]
df
```

```python
plt.figure(figsize=(20,20),facecolor='yellow')
number=1
for i in numerical_object:
    if number<=6:
        ax=plt.subplot(2,3,number)
        sns.distplot(df[i])
        plt.xlabel(i,fontsize=10)
    number+=1
plt.show()                                                                                       
```

```python
df.drop('Loan_ID',axis=1,inplace=True)
```

```python
df
```

```python
# Encoding the data 

from sklearn.preprocessing import LabelEncoder

lab_end=LabelEncoder()
df['Gender']=lab_end.fit_transform(df['Gender'])
df['Married']=lab_end.fit_transform(df['Married'])
df['Education']=lab_end.fit_transform(df['Education'])
df['Self_Employed']=lab_end.fit_transform(df['Self_Employed'])
df['Property_Area']=lab_end.fit_transform(df['Property_Area'])                                    
df
```

```python
df.info()
```

```python
# convert string categories into number using replace function 

df['Loan_Status']=df['Loan_Status'].replace({'Y':1,'N':0})
df
```

```python
# Check data balanced or not

df['Loan_Status'].value_counts()
```

```python
x=df.drop(['Loan_Status'],axis=1)
y=df.Loan_Status
```

```python
# Use sampling technique to convert imbalaned data into balanced

from imblearn.over_sampling import SMOTE

sm=SMOTE(random_state=40,k_neighbors=5,sampling_strategy=0.70)
x_samp,y_samp=sm.fit_resample(x,y)
```

```python
from collections import Counter

print("The frequecy of class before fit : ",format(Counter(y)))
print("The frequecy of class after fit : ",format(Counter(y_samp)))
```

```python
x_samp['Loan_Status']=y_samp
x_samp
```

```python
# Check the corelation and relationship b/w data (feature and Target)

plt.figure(figsize=(25,10))
sns.heatmap(x_samp.corr().abs(),annot=True)
plt.show()

# In this heapmap we can see no corelation is there between the features
```

```python
X=x_samp.drop(['Loan_Status'],axis=1)
Y=x_samp.Loan_Status
```

```python
# Now we check the relation b/w Input and Output Variable

plt.figure(figsize=(30,25),facecolor='yellow')
number=1
for i in X:
    if number<=11:
        ax=plt.subplot(2,6,number)
        plt.scatter(X[i],Y)
        plt.xlabel(i,fontsize=15)
        plt.ylabel('Loan_Status',fontsize=25)
    number+=1
plt.show()

# After go through the graph we analyse every feature has a relation with target so we can not remove any feature data on the 
# basis of assumption untill we are not sure.
```

```python
# Now standarised the data means take all data on one scale.

from sklearn.preprocessing import StandardScaler

x_scaler=StandardScaler()
x_scaler=x_scaler.fit_transform(X)
x_scaler
```

```python
# Here we use PCA Technique which convert feature into dimentions and select required dimentions and remaining will be removed.
```

```python
from sklearn.decomposition import PCA

pca=PCA()
pca.fit_transform(x_scaler)
```

```python
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.grid(axis='both')
plt.xticks(range(1,11,2))
plt.xlabel('Principle Components')
plt.ylabel('Varience Covered')
plt.title('PCA')
plt.show()
```

```python
pca=PCA(n_components=9) # here we select required dimensions number
new_comp=pca.fit_transform(x_scaler)
princi_comp=pd.DataFrame(new_comp)
princi_comp
```

```python
from sklearn.metrics import roc_curve,auc

data = {'Knn Model': knn,'Linear_Regression':lm,'Decision Tree':dec_tree,'Random_Forest':ran_fort,
                     'Support Vector Machines':svc}
data
```

```python
plt.figure(figsize=(5,6))
for data_name,data_object in data.items():
    data_object.fit(x_train,y_train)
    y_pred=data_object.predict(x_test)
    fpr,tpr,thrashold=roc_curve(y_test,y_pred)
    roc_auc=auc(fpr,tpr)
    plt.plot(fpr,tpr,label='{} (AUC= {:.2f}'.format(data_name,roc_auc))
    
plt.plot([0,1],[0,1],linestyle='--',color='red')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristics (ROC) Curve')
plt.legend(loc='lower right')
    
#show the plot
plt.show()       
```
