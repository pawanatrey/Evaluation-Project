```python
# Split Numerical data and Categorical Data
# It will tell these are the categorical column in dataset

categorical_object=[]
for i in df.dtypes.index:
    if df.dtypes[i]=='object':
        categorical_object.append(i)      
print("Categorical Data",categorical_object)    
```

```python
# It will tell these are the numerical column in dataset

numerical_object=[]
for i in df.dtypes.index:
    if (df.dtypes[i]!='object'):
        numerical_object.append(i) 
print("Numerical Data\n",numerical_object)          
```

```python
# Here we check how data is distributed, it is normally distributed or not.

# After go through the dataset we assume something that skewness and outliers is there so now we analyse the data to check 
# howmuch our assumption is correct.

plt.figure(figsize=(20,15),facecolor='yellow')
number=1
for i in numerical_object:
    if number<=6:
        ax=plt.subplot(2,3,number)
        sns.distplot(df[i])
        plt.xlabel(i,fontsize=10)
    number+=1
plt.show()                                                                                       
```

```python
plt.figure(figsize=(20,15),facecolor='yellow')
number=1
for i in numerical_object:
    if number<=6:
        ax=plt.subplot(2,3,number)
        sns.boxplot(df[i])
        plt.xlabel(i,fontsize=10)
    number+=1
plt.show()
```

```python
# Find the standard number of outlier and remove it

from scipy.stats import zscore

z_score=zscore(df[['Age', 'Fnlwgt', 'Education_num', 'Capital_gain', 'Capital_loss', 'Hours_per_week']])
data=np.abs(z_score)
filter_data=(data<3).all(axis=1)
df=df[filter_data]
df
```

```python
plt.figure(figsize=(20,20),facecolor='yellow')
number=1
for i in numerical_object:
    if number<=6:
        ax=plt.subplot(2,3,number)
        sns.distplot(df[i])
        plt.xlabel(i,fontsize=10)
    number+=1
plt.show()                                                                                       
```

```python
# Encoding the data 

from sklearn.preprocessing import LabelEncoder

for i in categorical_object:
    lab_end=LabelEncoder()
    df[i]=lab_end.fit_transform(df[i])
df.head()
```

```python
# Now we check the relation b/w Input and Output Variable

plt.figure(figsize=(30,25),facecolor='yellow')
number=1
for i in X:
    if number<=14:
        ax=plt.subplot(3,5,number)
        plt.scatter(X[i],Y)
        plt.xlabel(i,fontsize=15)
        plt.ylabel('Income',fontsize=25)
    number+=1
plt.show()

# After go through the graph we analyse every feature has a relation with target so we can not remove any feature data on the 
# basis of assumption untill we are not sure.
```

```python
# Select the feature which have most significant relation with the target variable.
from sklearn.feature_selection import SelectKBest,f_classif

best_features= SelectKBest(score_func=f_classif,k=10)                                                     
fit=best_features.fit(X,Y)                           
                                                                                                 
df_scores=pd.DataFrame(fit.scores_)                  
df_columns=pd.DataFrame(X.columns)

# concatenate dataframes

features_score=pd.concat([df_columns,df_scores],axis=1)
features_score.columns= ['Features_Name','Score']
print(features_score.nlargest(10,'Score')) 
```

```python
from sklearn.metrics import roc_curve,auc

data = {'Knn Model': knn,'Linear_Regression':lm,'Decision Tree':dec_tree,'Random_Forest':ran_fort,
                     'Support Vector Machines':svc}
data
```

```python
plt.figure(figsize=(5,6))
for data_name,data_object in data.items():
    data_object.fit(x_train,y_train)
    y_pred=data_object.predict(x_test)
    fpr,tpr,thrashold=roc_curve(y_test,y_pred)
    roc_auc=auc(fpr,tpr)
    plt.plot(fpr,tpr,label='{} (AUC= {:.2f}'.format(data_name,roc_auc))
    
plt.plot([0,1],[0,1],linestyle='--',color='red')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristics (ROC) Curve')
plt.legend(loc='lower right')
    
#show the plot
plt.show()       
```

```python
# Now in this ROC AUC metric we can see Random Forest Model make the highest curve so we will select RFM to this problem.
```
