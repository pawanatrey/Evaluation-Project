```python
# After go through the dataset we assume something so now we analyse the data to check howmuch i assumption is correct
# Here we check how data is distributed, it is normally distributed or not.
plt.figure(figsize=(20,20),facecolor='yellow')
number=1
for i in df:
    if number<=17:
        ax=plt.subplot(4,5,number)
        sns.distplot(df[i])
        plt.xlabel(i,fontsize=10)
    number+=1
plt.show()                                                                                       
```

```python
plt.figure(figsize=(20,20),facecolor='yellow')
number=1
for i in df:
    if number<=17:
        ax=plt.subplot(4,5,number)
        sns.boxplot(df[i])
        plt.xlabel(i,fontsize=10)
    number+=1
plt.show()
```

```python
# Find the standard number of outlier and remove it

from scipy.stats import zscore

z_score=zscore(df[['R','ERA','SHO','SV','E']])
data=np.abs(z_score)
filter_data=(data<3).all(axis=1)
df=df[filter_data]
df
```

```python
df.shape
```

```python
# Again check the distribution of data after remove the outliers

plt.figure(figsize=(25,10),facecolor='yellow')
number=1
for i in df[['R','ERA','SHO','SV','E']]:
    if number<=5:
        ax=plt.subplot(1,5,number)
        sns.distplot(df[i])
        plt.xlabel(i,fontsize=10)
    number+=1
plt.show()
```

```python
# Separate Input and Output Variable

x=df.drop(['W'],axis=1)
y=df.W
```

```python
# Now we check the relation b/w Input and Output Variable

plt.figure(figsize=(30,20),facecolor='yellow')
number=1
for i in x:
    if number<=16:
        ax=plt.subplot(4,5,number)
        plt.scatter(x[i],y)
        plt.xlabel(i,fontsize=25)
        plt.ylabel('W',fontsize=25)
    number+=1
plt.show()

# After go through the graph we analyse every feature has a relation with target so we can not remove any feature data on the 
# basis on assumption untill we are not sure.
```

```python
# Check the corelation and relationship b/w data (feature and Target)

plt.figure(figsize=(20,15))
sns.heatmap(df.corr().abs(),annot=True)
plt.show()

# In this heapmap we can see there is high corelation b/w RA & ER, RA & ERA, ER & ERA but they have a strong relation with 
# target so before remove any feature we reconfirm or collect more proff.
```

```python
# Now standarised the data means take all data on one scale.

from sklearn.preprocessing import StandardScaler

x_scaler=StandardScaler()
x_scaler=x_scaler.fit_transform(x)
x_scaler
```

```python
# check multicolinarity if vif < 10 then no multicolinarity

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif=pd.DataFrame()
vif['Vif']=[variance_inflation_factor(x_scaler,i) for i in range(x_scaler.shape[1])]
vif['Feature']=x.columns
vif
```

```python
X=x.drop(['ER'],axis=1)
```

```python
# Now standarised the data means take all data on one scale.

from sklearn.preprocessing import StandardScaler

x_scaler=StandardScaler()
x_scaler=x_scaler.fit_transform(X)
x_scaler
```

```python
# Split the dataset into Training and Test Data
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x_scaler,y,test_size=.25,random_state=40)
```

```python
# As it is Regression Problem so we use Regression Models
# Linear Regression Model

from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
# Now we check how much linear Regression model understand the trained data

lin_reg.score(x_train,y_train)
```

```python
# Now we check how much linear Regression model understand the test Data

lin_reg.score(x_test,y_test)
```

```python
# Now we evaluate the model using metrics that how model is performing.

from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
y_pred=lin_reg.predict(x_test)
y_pred
```

```python
r1=r2_score(y_test,y_pred)
r1
```

```python
mae1=mean_absolute_error(y_test,y_pred)
```

```python
# In above DataFrame we can see Linear Regression model performing better than other models so we use hyperparameter tunning
# for Linear Regression.
# For standard linear regression, there are no hyperparameters to tune. However, when we extend linear regression to include 
# regularization techniques such as Ridge Regression, Lasso Regression.

from sklearn.linear_model import Ridge,RidgeCV
```

```python
ridge=Ridge()
ridge.fit(x_train,y_train)
```

```python
parameter={'alpha':[0.0001,0.001,0.01,0.1,1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(ridge,parameter,cv=5)
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
# Final Model

final_model=Ridge(alpha=1)
final_model.fit(x_train,y_train)
```

```python
predd=final_model.predict(x_test)
predd
```
